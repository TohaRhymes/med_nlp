## Selection part

```python
from utils import key2word, key2ruword, word2key
import pandas as pd
import numpy as np
from collections import defaultdict
import json

from copy import deepcopy

from typing import List


cols = ["category_meta", "category_fg", "category_ukb"]
code_cols = {
    "category_meta": "fg_code",
    "category_fg": "fg_code",
    "category_ukb": "ukb_code",
}
n_cols = {
    "category_meta": "fg_significant",
    "category_fg": "ukb_significant",
    "category_ukb": "meta_significant",
}
links = {
    "category_meta": "https://public-metaresults-fg-ukbb.finngen.fi/pheno/",
    "category_fg": "https://risteys.finregistry.fi/endpoints/",
    "category_ukb": "https://biobank.ndph.ox.ac.uk/showcase/field.cgi?id=",
    
}
links_pheno = {
    "category_meta": "https://risteys.finregistry.fi/endpoints/",
    "category_fg": "https://risteys.finregistry.fi/endpoints/",
    "category_ukb": "",
    
}


similarity = pd.read_csv('data/final_similarity.tsv', sep='\t')
similarity['similarity_category_fg'] = similarity[['similarity_category_fg', 'similarity_fg_name']].max(axis=1, skipna=True)
similarity['similarity_category_ukb'] = similarity[['similarity_description_more_ukb', 'similarity_description_ukb']].max(axis=1, skipna=True)
similarity['category_fg'] = similarity['fg_name'] + ": " + similarity['category_fg']
similarity['category_ukb'] = similarity['description_ukb'] + ": " + similarity['description_more_ukb']
similarity.drop(['similarity_fg_name', 'fg_name', 'similarity_description_more_ukb', 'similarity_description_ukb', 'description_more_ukb', 'description_ukb'] , axis=1, inplace=True)
similarity['en'] = similarity['en'].apply(lambda x: word2key[x])

similarity.sample(10)
```

```python
def return_similarities(codes:List[str],
                        similarity_cutoff:float = 0.92,
                        K_cutoff:int = 5,
                        use_similarity:bool = False):
    similarity_codes = deepcopy(similarity[similarity.en.isin(codes)])

    phenotypes = defaultdict(list)
    gwases = defaultdict(list)

    for col in cols:
        cur_similarity = similarity_codes[similarity_codes[col].notnull()].sort_values(
            f"similarity_{col}", ascending=False
        )
        if use_similarity:
            cur_similarity = cur_similarity[cur_similarity[f"similarity_{col}"]>similarity_cutoff]
        else:
            cur_similarity = cur_similarity.head(K_cutoff)
        cur_similarity['link'] = links[col] + cur_similarity[code_cols[col]]
        cur_similarity['significant'] = cur_similarity[n_cols[col]]
        needed_cols = [code_cols[col], col, 'en', f"similarity_{col}", "link", "significant"]
        grouped_corrected = cur_similarity[needed_cols]

        gwases[col] = grouped_corrected.to_dict()
        phenotypes[col] =  [(links_pheno[col] + i, i) for i in list(set(grouped_corrected[col]))]

    phenotypes = dict(phenotypes)
    gwases = dict(gwases)

    response = {
                'phenotypes': phenotypes,
                'gwases': gwases,
            }

    return response
```

```python
codes = ['type_', 'collection_pop', 'collection_spon']
similarity_cutoff = 0.92
K_cutoff = 5
use_similarity = False

response = return_similarities(codes = codes,
                    similarity_cutoff = similarity_cutoff,
                    K_cutoff = K_cutoff,
                    use_similarity = use_similarity)
response
# return Response(response)
```
