```python
import pandas as pd
```

## first three

```python
data1 = pd.read_csv('data/similarity_category_meta.csv').drop(['Unnamed: 0'], axis = 1)
data2 = pd.read_csv('data/similarity_fg_name.csv').drop(['Unnamed: 0'], axis = 1)[['similarity_fg_name',]]
data3 = pd.read_csv('data/similarity_category_fg.csv').drop(['Unnamed: 0'], axis = 1)[['similarity_category_fg',]]

display(data1.sample(7))
display(data2.sample(7))
display(data3.sample(7))
print(data1.shape)
print(data2.shape)
print(data3.shape)
```

```python
data123 = data1.merge(
    data2, 
    left_index=True, 
    right_index=True, 
    how='outer'
).merge(
    data3, 
    left_index=True, 
    right_index=True, 
    how='outer'
)
print(data123.shape)
data123.sample(7)
```

## Add fourth

```python
data45 = []
for col in ['similarity_description_more_ukb', 'similarity_description_ukb']:
    data_p1 = pd.read_csv(f"data/{col}_0.csv").drop(["Unnamed: 0"], axis=1)
    data_p2 = pd.read_csv(f"data/{col}_1.csv").drop(["Unnamed: 0"], axis=1)
    data_p3 = pd.read_csv(f"data/{col}_2.csv").drop(["Unnamed: 0"], axis=1)
    data = pd.concat([data_p1, data_p2, data_p3]).reset_index().drop(["index"], axis=1)[[col,]]
    data45.append(data)


print(data45[0].shape)
print(data45[1].shape)
display(data45[1].sample(7))
display(data45[0].sample(7))
```

## Merge all

```python
data_all = data123.merge(
    data45[0], 
    left_index=True, 
    right_index=True, 
    how='outer'
).merge(
    data45[1], 
    left_index=True, 
    right_index=True, 
    how='outer'
)
print(data_all.shape)
data_all.sample(7)
```

```python
data_all[['fg_significant','ukb_significant','meta_significant']] = 0
data_all
```

```python
data_all.to_csv('data/final_similarity.tsv', sep='\t', index=False)
```
